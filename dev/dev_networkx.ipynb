```python
from pyvis.network import Network

# Create a binary tree with NetworkX
B = nx.balanced_tree(r=2, h=3)
node_info = {
    0: {"id": "1", "type": "A", "value": 99},
    1: {"id": "2", "type": "B", "value": 50},
    2: {"id": "3", "type": "B", "value": 75},
    3: {"id": "4", "type": "C", "value": 10},
    4: {"id": "5", "type": "C", "value": 25},
    5: {"id": "6", "type": "C", "value": 30},
    6: {"id": "7", "type": "C", "value": 45},
    7: {"id": "8", "type": "C", "value": 45},
    8: {"id": "9", "type": "C", "value": 45},
    9: {"id": "10", "type": "C", "value": 45},
    10: {"id": "11", "type": "C", "value": 45},
    11: {"id": "12", "type": "C", "value": 45},
    12: {"id": "13", "type": "C", "value": 45},
    13: {"id": "14", "type": "C", "value": 45},
    14: {"id": "15", "type": "C", "value": 45},
    15: {"id": "16", "type": "C", "value": 45},
}

titles = {
    node: f'id: {info["id"]}' for node, info in node_info.items()
}

nx.set_node_attributes(B, titles, 'title')


# Create a pyvis Network object
net = Network(notebook=True, directed=True, cdn_resources='in_line')

# Add physics layout options for a hierarchical view
net.set_options("""
var options = {
  "layout": {
    "hierarchical": {
      "enabled": true,
      "levelSeparation": 150,
      "nodeSpacing": 100,
      "treeSpacing": 200,
      "direction": "UD",
      "sortMethod": "hubsize"
    }
  },
  "physics": {
    "enabled": false
  }
}
""")


# Load the NetworkX graph into pyvis
net.from_nx(B)

# Show the interactive graph in your browser
net.show("interactive_tree.html")
```
